```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
ori_df = pd.read_csv("./final_3.csv")
ori_df
```

```python
df = ori_df.copy()
df
```

```python
df.set_index('id', inplace=True)
df
```

```python
df = df.mask(df.applymap(str).eq('[]'))
df
```

```python
df.replace('', np.NaN, inplace=True)
df
```

```python
df.shape
```

## missing value 佔比

```python
ans = df.isna().sum().sum()/(df.shape[0]*df.shape[1])
print(ans)
```
